```python
def initialize_vectors(topElist, toppxlist, toppylist, toppzlist):
    vector_list = []
    num_events = len(topElist)
    for event in range(num_events):
        vector_list.append(vector.obj(px=toppxlist[event], py=toppylist[event], pz=toppzlist[event], E=topElist[event]))
    return vector_list
```

```python
vectorLlist=initialize_vectors(topELlist, toppxLlist, toppyLlist, toppzLlist)
```

```python
#print(vectorLlist)
```

```python
def plot_mass_from_vectors(title, vectors):
    m_list = []
    for vec in vectors:
        m_list.append(vec.m)
    plt.figure()
    plt.title(title)
    plt.hist(numpy.ravel(m_list),bins=30)
    plt.savefig(title+".png")
    plt.show()
    plt.close()
```
